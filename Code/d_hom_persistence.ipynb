## We calculate the persistent homology of our data to know its shape

```r
library('MASS')
#library("DescTools")
library('TDA')
source('../../Tesis/Code/R/plot.diagram.R')
library("RcppCNPy")
```

```r
#Geodesic distance (from c_distance6) highlights better the result
diss=npyLoad('../Data/dist/all_l1-thres-geo.npy')
```

```r
#------------------------------------------------------------------------
#-------------------------generate cycles and diagram---------------------------------
maxdim=1
#maxscale=max(diss[,1])
maxscale=250
List=ripsDiag(X=diss, maxdim, maxscale, dist="arbitrary", library="Dionysus",location=TRUE)
Diag=List$diagram 
cycles=List$cycleLocation
```

```r
m=nrow(Diag)
for (k in 2:nrow(Diag)){#Number of 0-classes
	if (Diag[k,1]!=0){
		m=k-1 #(The row whit dimension 1)-1
		break
	}
}
```

```r
max=maxscale
#max=Diag[1:m,3] #Maxscale
ind=4*0:(max/4) #Divide in 4 maxscale
values=sort(Diag[1:m,3]) #Death points of 0-classes sorted
l=length(values)
real_max=values[l-1] #After it there is only one cluster
sizes=integer()
for (r in ind){#Number of clusters for each r in ind
	for(i in 1:l){
		if(values[i]>r){break}	
	}
	sizes=c(sizes,toString(m-i+1))	#Is string value	
}
```

```r
#nombre=paste("../Imagenes/thesis/barcode.png",sep="")#nombre del archivo
name_plot=paste(" distance")#title plot
#png(nombre,width=500,height=500)
par(mar=c(5,5.5,4,2)+0.1,cex.axis=1.4,cex.main=1.5,cex.lab=1.4,las=1) #cex = size , las = orientation labels (1=horizontal)
plot(Diag,barcode=F,)#barcode
#title(name_plot, line = 4)
#dev.off()
```

```r
nombre=paste("barcode.pdf",sep="")#nombre del archivo
name_plot=paste(" distance")#title plot
#pdf(nombre,width=14,height=6)
par(mar=c(5,5.5,4,2)+0.1,cex.axis=1.4,cex.main=1.5,cex.lab=1.4,las=1) #cex = size , las = orientation labels (1=horizontal)
plot(Diag,barcode=T,dimension=1,xlab="Ratios")#barcode
#title(name_plot, line = 4)
axis(2,at=m/2,labels=expression("H"[0]),cex=1.4) #left axis
axis(1,at=ind) #below axis
axis(3,at=ind,labels=sizes) #above axis
for (r in ind){
    lines(c(r,r),c(-18,as.integer(sizes[1])+20),lty=2,col="gray")#dash lines
}
 mtext("Number of",2, line=0,at=(as.integer(sizes[1])+104),cex=1.3)
 mtext("clusters",2, line=0.4,at=(as.integer(sizes[1])+64),cex=1.3)
#dev.off()
```
